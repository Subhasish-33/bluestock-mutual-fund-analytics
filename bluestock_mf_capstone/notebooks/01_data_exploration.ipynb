# Data Exploration & Quality Check

This notebook loads all CSV files from `data/raw/` and performs initial data quality checks.

## Process:
1. Load all CSV files
2. Inspect shape and data types
3. Check for null values
4. Display sample rows
5. Document anomalies

```python
import pandas as pd
import os
import glob

# Load all CSV files from data/raw/
raw_path = "../data/raw/"
csv_files = glob.glob(raw_path + "*.csv")

datasets = {}
for f in csv_files:
    name = os.path.basename(f).replace(".csv", "")
    df = pd.read_csv(f)
    datasets[name] = df
    print(f"\n=== {name} ===")
    print("Shape:", df.shape)
    print("Dtypes:\n", df.dtypes)
    print(df.head(3))
    print("Nulls:\n", df.isnull().sum())
```
